## Method 1: MLP Regressor

```python
from sklearn.neural_network import MLPRegressor as MLPR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error as mape
rnd = 61353
```

```python
X = traindata
y = traintarget
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=rnd)
```

###  MLPR1

```python
mlpr = MLPR(hidden_layer_sizes = (50,50,10,),random_state = rnd,max_iter = 50, verbose = 1)
mlpr.fit(X_train,y_train)
```

```python
y_pred = mlpr.predict(X_test)
```

```python
print('Score: ',mape(y_test,y_pred))
```

### MLPR2

```python
mlpr2 = MLPR(hidden_layer_sizes = (50,50,25,10,),random_state = rnd,max_iter = 50, verbose = 1)
mlpr2.fit(X_train,y_train)
```

```python
y_pred = mlpr2.predict(X_test)
print('Score: ',mape(y_test,y_pred))
```

## Optimizing RFR

```python
ests = [10,20,30,40,50,100]
deps = [100,200,300]
nests = []
maxdeps = []
score = []
```

```python
for e in ests:
    for d in deps:
        nests.append(e)
        maxdeps.append(d)
        rfr = RandomForestRegressor(n_estimators = e, max_depth = d,verbose = 1)
        rfr.fit(X_train,y_train)
        y_pred = rfr.predict(X_test)
        s = mape(y_test,y_pred)
        print('Nest = ',e,', Max Dep. = ',d,'; Score = ',s)
        score.append(s)
        
```

Best score n_estimators = 100, max_depth = 100, score = ~0.01608.

Training on full data set.

## Attempt to separate by turbine_id

```python
traintypes = trainid
testtypes = testid
```

```python
uniquetraintypes = np.unique(traintypes)
uniquetesttypes = np.unique(testtypes)
print(uniquetraintypes)
```

```python
for i in uniquetesttypes:
    if i not in uniquetraintypes:
        print('Unmatched type found')
```

```python
rfrs = {}
for id in uniquetraintypes:
    print('Now training: ',id)
    rfrs[id] = RandomForestRegressor(n_estimators = 100, max_depth = 100, verbose = 2)
    rfrs[id].fit(X[traintypes == id],y[traintypes == id])
    

```

```python
sub_target = np.zeros(len(testdata))
```

```python
for id in uniquetesttypes:
    sub_target[testtypes == id] = rfrs[id].predict(testdata[testtypes == id])
```

```python
np.savetxt("submission_idwise.csv", sub_target, delimiter=",")
```

id-wise RFR gives mape = 0.013000

## Keras

```python
from tensorflow.python.keras.models import Model
from tensorflow.python.keras.layers import Input,Dense
```

```python
input = Input(shape = (X.shape[1],))
layer = Dense(np.sqrt(len(X)),activation = 'relu')(input)
layer = Dense(np.sqrt(np.sqrt(len(X))),activation = 'relu')(layer)
layer = Dense(10,activation = 'relu')(layer)
output = Dense(1,activation = 'relu')(layer)
model = Model(input,output)
```

```python
model.compile('adam', loss = 'mean_absolute_percentage_error')
model.summary()
```

```python
hist_reg = model.fit(X,y,epochs = 100, batch_size = 32, shuffle = True, validation_split = 0.1)
```

```python
preds = model.predict(testdata)
print(preds.shape)
```

```python
model.evaluate(X,y)
```

### id-wise keras

```python
from tensorflow.keras.models import Sequential
```

```python
lrs = {}
for id in uniquetraintypes:
    print('Now training: ',id)
    lrs[id] = Sequential()
    lrs[id].add(Dense(50,input_dim=X.shape[1],activation = 'relu'))
    lrs[id].add(Dense(100,activation = 'relu'))
    lrs[id].add(Dense(50,activation = 'relu'))
    lrs[id].add(Dense(1,activation = 'relu'))
    
    lrs[id].compile('adam', loss = 'mean_absolute_percentage_error',metrics = ['mean_absolute_percentage_error'])
    lrs[id].summary()
    
    
    lrs[id].fit(X[traintypes == id],y[traintypes == id],epochs = 50, batch_size = 32, shuffle = True, validation_split = 0.1)
    
sub_target = np.zeros(len(testdata))
```

```python
for id in uniquetesttypes:
    sub_target[testtypes == id] = lrs[id].predict( testdata[testtypes == id])
    
np.savetxt("submission_kerasidwise.csv", sub_target, delimiter=",")
```
